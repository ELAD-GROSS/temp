```python
from simulator import LoadBalancer


def dry_b_results():
    times_list = [10, 100, 500, 1000, 2000, 2500]
    results = dict()
    test_amount = 10

    for t in times_list:
        load_balancer = LoadBalancer(T=t, N=1, prob=[1], lambada=10, queue_sizes=[5], rates=[12])
        tot_wait_time = 0
        for _ in range(test_amount):
            _, _, _, _, tot_serving_time = load_balancer.simulate_load_balancer()
            tot_wait_time += tot_serving_time
        results[t] = tot_wait_time / test_amount
    return results
```

```python
results = dry_b_results()
```

```python
import matplotlib.pyplot as plt

def draw_graph(results):

    x_values = list(results.keys())
    y_values = list(results.values())

    # Create an empty figure and axis
    fig, ax = plt.subplots()

    # Plot the x-y points
    ax.plot(x_values, y_values, 'bo')

    # Connect the points with lines
    for i in range(len(x_values) - 1):
        ax.plot([x_values[i], x_values[i+1]],
                [y_values[i], y_values[i+1]], 'b-')

    # Set the labels for x-axis and y-axis
    ax.set_xlabel('waiting time')
    ax.set_ylabel('simulator run time')

    # Set the title of the graph
    ax.set_title('Avg of waiting time by simulator run time')

    # Show a grid
    ax.grid(True)

    # Show the graph
    plt.show()
```
